```python
import os

# working_dir = './expts_result/'

working_dir = f'./expts_result/'


# define exprimental settings
n_runs = 10
dim = 5
init_eval_num = 10
target_func = sphere_function
surrogate = SingleTaskGP

dim_subdir = f'{dim}_dim'

save_dir = os.path.join(working_dir, dim_subdir)

if not os.path.exists(save_dir):
    os.makedirs(save_dir)


accumulated_loss = 0
target_value = 0

bo_num = 10

for i in range(bo_num):
    res_df = run_bo(n_runs, dim, init_eval_num, target_func, surrogate)
    accumulated_loss += abs(target_value - res_df['最適値'].min())
    save_path = os.path.join(save_dir, f'opt_{i+1}.csv')
    res_df.to_csv(save_path)

print(f'総損失: {accumulated_loss}')

# 設定を辞書として定義
settings = {
    'n_runs': n_runs,
    'dim': dim,
    'init_eval_num': init_eval_num,
    'target_func': str(target_func),
    'surrogate': str(surrogate),
    'accumulated_loss': surrogate,
    'target_value': target_value,
    'bo_num': bo_num
}

# 設定をテキストファイルに保存
settings_file_path = os.path.join(save_dir, 'settings.txt') 
with open(settings_file_path, 'w') as f:
    for key, value in settings.items():
        f.write(f'{key}: {value}\n')
```

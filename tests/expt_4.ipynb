```python
import matplotlib.pyplot as plt
from matplotlib import cm
from gpytorch.kernels import RBFKernel

fig = plt.figure(figsize=(10, 10))

xs = torch.linspace(-5, 5, steps=100).requires_grad_(True)
ys = torch.linspace(-5, 5, steps=100).requires_grad_(True)
x, y = torch.meshgrid(xs, ys, indexing='xy')

tmp = torch.cat([x.ravel().reshape(-1, 1), y.ravel().reshape(-1, 1)], 1)
origin = torch.Tensor([[0., 0.]]).requires_grad_(True)

f = RBFKernel()
f.lengthscale = 1.
z = f(tmp, origin).evaluate().reshape(100, 100)

z.backward(torch.ones_like(z), retain_graph=True)
grad_x = xs.grad.numpy()
grad_y = ys.grad.numpy()

ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x.detach().numpy(), y.detach().numpy(), z.detach().numpy(), rstride=2, cstride=2, alpha=0.9)

plt.show()
```
